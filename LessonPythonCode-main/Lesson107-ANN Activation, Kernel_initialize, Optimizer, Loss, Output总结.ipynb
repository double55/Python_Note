### （2）Loss functions

classification:
1. binary cross entropy -- Binary Classification
2. categorical cross entropy -- softmax(Multi-Classification) -- ie 001,010,100
3. sparse categorical cross entropy -- softmax(Multi-Classification) -- ie 0,1,2
-If targets are one-hot encoded, use categorical_crossentropy, But if targets are integers, use sparse_categorical_crossentropy.

Regression:
1. mean absolute error
2. mean squared error
3. root mean squared error

### (3) Optimizer

how to reduce errors which use optimizers:
1. Gradient Descent - takes all records and forward, computing loss and backprobagation -- resource issue
2. SGD - (Stochastic gradient descent) 1 record and forward, computing loss and backprobagation -- slowing training
3. Mini batch SGD - Forward, backward, propogation (Noise convergence will take time)
4. SGD with momentum
5. Adagrad
6. Adadelta & RMSprop
7. Adam (momentum + RMSprop) -- best optimizer

### (4) Kernel_initialize

why does exploding gradient problem happen?

weight initializtion techniques
Xavier Glorot, He init

-He + relu 

It is a good practice to use ‘relu‘ activation with a ‘he_normal‘ weight initialization. This combination goes a long way to overcome the problem of vanishing gradients when training deep neural network models

-Xavier Glorot + sigmoid or tanh 

He initialization works better for layers with ReLu activation.

Xavier initialization works better for layers with sigmoid activation.
