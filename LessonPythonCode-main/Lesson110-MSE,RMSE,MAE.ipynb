```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([10,20,30,40,50,60])
y=np.array([15,45,50,70,110,130])
```

```python
linreg=LinearRegression()
```

```python
linreg
```

```python
x
```

```python
X=x.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
linreg.fit(X,y)
```

```python
y_predict=linreg.predict(X)
```

```python
y_predict
```

```python
plt.scatter(X,y)
#plt.plot(X,y,'Blue')
plt.plot(X,y_predict,'Red')
plt.show()
```

```python
print(linreg.coef_)
```

```python
print(linreg.intercept_)
```

```python
linreg.predict([[75]])
```

```python
from sklearn import metrics

#print MAE
print(metrics.mean_absolute_error(y, y_predict))

#print MSE
print(metrics.mean_squared_error(y, y_predict))

#print RMSE
print(np.sqrt(metrics.mean_squared_error(y, y_predict)))

#print R2
print(metrics.r2_score(y,y_predict))
```

```python
# calculate manually
d = y - y_predict
mse_f = np.mean(d**2)
mae_f = np.mean(abs(d))
rmse_f = np.sqrt(mse_f)
r2_f = 1-(sum(d**2)/sum((y-np.mean(y))**2))

print("Results by manual calculation:")
print("MAE:",mae_f)
print("MSE:", mse_f)
print("RMSE:", rmse_f)
print("R-Squared:", r2_f)
```

MSE is highly biased for higher values. RMSE is better in terms of reflecting performance when dealing with large error values. RMSE is more useful when lower residual values are preferred.

go for Mean Absolute Error when see Any outlier as it is robust to outlier if not then go for Mean square error as it penalizes the model for error by squaring it .

If outliers in dataset we usually go with MAE else we can use MSE
